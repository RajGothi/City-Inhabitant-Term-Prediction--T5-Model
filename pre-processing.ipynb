```python
import matplotlib.pyplot as plt

display_threshold = 30

# Count the occurrences of each class
suffix_counts = df['Suffix'].value_counts()

# Identify classes to group as "Others"
classes_to_group = suffix_counts.index[display_threshold:]

# Group the classes as "Others" and sum their counts
suffix_counts['Others'] = suffix_counts[classes_to_group].sum()
suffix_counts.drop(classes_to_group, inplace=True)
plt.figure(figsize=(10, 7)) 
# Create a bar plot
plt.bar(suffix_counts.index, suffix_counts.values)
# Add labels and a title
plt.xticks(rotation=45)  
plt.xlabel('Suffix')
plt.ylabel('Count')
plt.title('Count of Each Suffix Value (with "Others")')

plt.savefig('suffix_counts_plot.png')

# Show the plot
plt.show()
```

```python
import re

all_city = []
term_city = []
for index,rows in df.iterrows():
    city = rows["City"]
    term = rows["Inhabitant_term"]
    city = re.sub(chars_to_ignore_regex, ' ', city)
    city = city.strip()
    city = city.lower()
        
    term = re.sub(chars_to_ignore_regex, ' ', term)
    term =term.strip()
    term=term.lower()

    city = city.replace('▁','')
    term = term.replace('▁','')
    all_city.append(city)
    term_city.append(term)
```

```python
new_df = pd.DataFrame({'city': all_city,'term': term_city,'country':df["Country"],'suffix':df['Suffix']})
new_df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import math

# Load your dataset into a DataFrame (assuming it's in a CSV file)
# df = pd.read_csv("your_dataset.csv")
df = new_df

# Get the unique suffixes
unique_suffixes = df['suffix'].unique()

# Initialize empty dataframes for training, validation, and test sets
train_df = pd.DataFrame()
val_df = pd.DataFrame()
test_df = pd.DataFrame()

min_samples = 10  

# Split the data for each suffix
for suffix in unique_suffixes:
    suffix_df = df[df['suffix'] == suffix]
    
    num_samples = len(suffix_df)
    print(suffix)

    if suffix not in select_suffix:
        train_df = train_df.append(suffix_df)
        continue  

    # if num_samples < min_samples:
    #     train_df = train_df.append(suffix_df)
    #     continue  

    print(num_samples)
    # Calculate the required proportion of samples for each split
    train_samples = min(math.floor(num_samples * 0.8), num_samples - 1)
    val_samples = min(math.floor(num_samples * 0.1), num_samples - train_samples - 1)
    # if val_samples==0:
    #     val_samples=1
    test_samples = num_samples - train_samples - val_samples
    print(train_samples,val_samples,test_samples)
    # Split the suffix_df into train, val, and test based on calculated proportions
    train, temp = train_test_split(suffix_df, train_size=train_samples, random_state=42)
    val, test = train_test_split(temp, train_size=val_samples, random_state=42)
    
    # Append the results to the respective dataframes
    train_df = train_df.append(train)
    val_df = val_df.append(val)
    test_df = test_df.append(test)

# Now you have train_df, val_df, and test_df with an approximate 80-10-10 split based on suffixes,
# while ensuring a minimum number of samples for each split
```
